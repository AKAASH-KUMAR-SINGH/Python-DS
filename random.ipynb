```python
from numpy import random
```

```python
# Generate a random integer from 0 to 100
x=random.randint(100)
print(x)
```

```python
# Generate a random float from 0 to 10
x=random.rand(1)
print(x)
```

```python
# Generate a array with size 7
x=random.randint(100,size=(7))
print(x)
```

```python
# Generate a 2-d array with size (3,4)
arr=random.randint(20,size=(3,4))
print(arr)
```

```python
# Generate a float valued array with size (3,4)
arr=random.rand(20)
print(arr)
```

```python
arr=random.rand(4,3)
print(arr)
```

```python
# choice() method is used to generate a random
# value from an array
arr=random.choice([2,3,1,4,2,1])
print(arr)
```

```python
# choice() method also allow you to return an array value.
# add size() parameter to specify the shape of the array
arr=random.choice([2,4,3,5,6,7,8,0],size=(2,5))
print(arr)
```

```python
arr=random.choice([3,4,5,6,2,1],size=(2,5))
print(arr)
```

----->>>>>RANDOM PERMUTATION<<<<<----

```python
# A permutation refers to an arrangement of element
    # numpy Random module provide two method
    #     shuffle()
    #     permutation
```

```python
import numpy as np
from numpy import random
arr=np.array([1,5,4,3,6,2])
random.shuffle(arr) # shuffle means changing arrangement in place.
print(arr)
```

```python
arr=np.array([2,4,3,1,5,6,8])
print(random.permutation(arr))
# The permutation() method return a rearrangement
# array 
```

---->>>> NORMAL (Gaussian) DISTRIBUTION<<<<-----

```python
# Use the random.normal() method to get a normal Data Distribution
# THere are Three parameter:
#     loc -(mean) where the peak of the bell exists.
#     scale -(Standard Deviation) how flat the graph distribution should be.
#     size -The shape of the returned array.
```

```python
arr=random.normal(size=(2,3))
print(arr)
```

```python
arr=random.normal(loc=1,size=(2,3))
print(arr)
```

```python
arr=random.normal(loc=2,size=(2,3),scale=2)
print(arr)
```

```python
sns.distplot(arr,hist=False)
plt.show()
```

```python
sns.distplot(arr)
plt.show()
```

```python
sns.distplot(random.normal(size=1000),hist=False)
plt.show()
```

```python
sns.distplot(random.normal(loc=1,scale=2,size=100),hist=False)
plt.show()
```
