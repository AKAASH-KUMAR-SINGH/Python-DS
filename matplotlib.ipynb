```python
# markersize or ms
xpoint=np.array([5,9,4,3,2,1,4,3,6,5])
ypoint=np.array([10,20,22,12,14,34,23,21,4,2])
plt.plot(xpoint,ypoint,'o:r',markersize=20)
```

```python
# markeredgecolor or mec
xpoint=np.array([5,9,4,3,2,1,4,3,6,5])
ypoint=np.array([10,20,22,12,14,34,23,21,4,2])
plt.plot(xpoint,ypoint,'o-r',ms=10,markeredgecolor='b')
```

```python
xpoint=np.array([5,9,4,3,2,1,4,3,6,5])
ypoint=np.array([10,20,22,12,14,34,23,21,4,2])
plt.plot(xpoint,ypoint,'o-r',markersize=10,markeredgecolor='k',markerfacecolor='b')
plt.show()
```

```python
xpoint=np.array([5,9,4,3,2,1,4,3,6,5])
ypoint=np.array([10,20,22,12,14,34,23,21,4,2])
plt.plot(xpoint,ypoint,'o:r',ms=5,mec='k',mfc='b')
plt.show()
```

```python
# linestyle   (ls)
xpoint=np.array([5,9,4,3,2,1,4,3,6,5])
ypoint=np.array([10,20,22,12,14,34,23,21,4,2])
plt.plot(xpoint,ypoint,'o:r',ms=10,mec='k',mfc='b',linestyle='-.')
plt.show()
```

```python
# Line-Color  'c'
xpoint=np.array([5,9,4,3,2,1,4,3,6,5])
ypoint=np.array([10,20,22,12,14,34,23,21,4,2])
plt.plot(xpoint,ypoint,'o:r',ms=5,mec='k',mfc='c',ls='dashdot',color='y')
plt.show()
```

```python
# Line-Width  (lw)
xpoint=np.array([5,9,4,3,2,1,4,3,6,5])
ypoint=np.array([10,20,22,12,14,34,23,21,4,2])
plt.plot(xpoint,ypoint,'o:r',ms=5,mec='k',mfc='b',linestyle='dotted',c='b',linewidth='5.20')
plt.show()
```

```python
# Set Line property for the grid
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'arial','color':'yellowgreen','size':12}
font2={'family':'serif','color':'green','size':10}
plt.title('Sport Watch Data',fontdict=font1,loc='left')
plt.xlabel('Average Pulse',fontdict=font2)
plt.ylabel('Calorie Burnage',fontdict=font2)
plt.grid(axis='x',color='blue',lw=1.5,ls='-.')
plt.plot(x,y,'o:r',ms=12,mfc='b',lw=3,c='y')
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y,c='green')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y,c='pink')
```

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,6])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,c=colors)
plt.show()
```

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,6])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,cmap='viridis',c=colors)
plt.show()
```

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,6])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,cmap='viridis',c=colors)
plt.colorbar()
plt.show()
```

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,6])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,cmap='viridis',c=colors,s=sizes,alpha=0.5)
plt.colorbar()
plt.show()

```
