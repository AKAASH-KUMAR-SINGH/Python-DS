```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.stripplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue='sex')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue='sex',palette='rocket_r')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue='sex',palette='rocket_r',linewidth=1)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue='sex',palette='rocket_r',linewidth=1,edgecolor='r')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,
              hue='sex',palette='rocket_r',
              linewidth=1,edgecolor='r',size=3,
              marker='<')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,
              hue='sex',palette='rocket_r',
              linewidth=1,edgecolor='r',size=3,
              marker='<',alpha=.7)
plt.show()
```

```python
sns.stripplot(x=var['total_bill'])
plt.show()
```

```python
sns.stripplot(y=var['total_bill'])
plt.show()
```

# Box Plot in seaborn

```python
var=sns.load_dataset('tips')
var
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var,hue='time')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var,hue='time',color='g')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var,hue='time',color='g',order=['Fri',"Sun",'Thur',"Sat"])
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var,hue='time',order=['Fri',"Sun",'Thur',"Sat"],showmeans=True,
            meanprops={'marker':'+','markeredgecolor':'b'})
plt.show()
```
