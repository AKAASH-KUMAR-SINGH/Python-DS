# Violin Plot in seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.violinplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=var,hue='time')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=var,linewidth=2)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=var,linewidth=2,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=['Dinner',"Lunch"],linewidth=2,palette='Dark2')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=['Dinner',"Lunch"],linewidth=2,palette='Dark2',color='r')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=var,hue='sex',split=True,scale='width')  
plt.show()
```

```python
sns.violinplot(x='total_bill',y='day',data=var,hue='sex')  
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=['Dinner',"Lunch"],inner='quart')     #{inner:"box","quartile",'point,stick,none}
plt.show()
```

```python
sns.violinplot(x=var['total_bill'])
plt.show()
```

```python
sns.violinplot(y=var['total_bill'])
plt.show()
```

# Pair Plot in seaborn

```python
var=sns.load_dataset('tips')
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue='sex')
plt.show()
```

```python
sns.pairplot(var,hue='sex',vars=['total_bill','tip'])
plt.show()
```

```python
sns.pairplot(var,hue='sex',vars=['total_bill','tip'],hue_order=['Female','Male'])
plt.show()
```

```python
sns.pairplot(var,hue='sex',vars=['total_bill','tip'],hue_order=['Female','Male'],palette='BuGn')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuGn',x_vars=['total_bill','tip'])
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],palette='BuGn',y_vars=['total_bill','tip'])
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'])
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],kind='reg')  #{kind=scatter,kde,hist,reg }
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],kind='reg',diag_kind='kde')
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],markers=['*','>'])
plt.show()
```
