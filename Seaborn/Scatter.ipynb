# Scatter plot in seaborn 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('penguins')
data
```

```python
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data)
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data,hue='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data,hue='sex',style='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data,hue='sex',
                style='sex',size='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data,hue='sex',
                style='sex',size='sex',sizes=(60,40))
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data,hue='sex',
                style='sex',size='sex',
                sizes=(60,40),palette='gist_ncar')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data,hue='sex',
                style='sex',size='sex',
                sizes=(60,40),palette='gist_ncar',
                alpha=.7)
plt.show()
```

```python
m={'Male':'*',"Female":'o'}
sns.scatterplot(x='bill_length_mm',
                y='bill_depth_mm',data=data,hue='sex',
                style='sex',size='sex',
                sizes=(60,40),palette='gist_ncar',
                markers=m)
plt.show()
```

```python
data=sns.load_dataset("anagrams")
data
```

```python
# removing the column from dataset
data=sns.load_dataset("anagrams")
x=data.drop(columns=['attnr'],axis=1).head(10)
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,vmin=0,vmax=14)
plt.show()
```

```python
sns.heatmap(x,vmin=0,vmax=10,cmap='gist_heat')
```

```python
sns.heatmap(x,vmin=0,vmax=10,cmap='PuOr',annot=True)
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='gist_heat',annot=True)
```

```python
ar=np.array([['a0','a1','a2','a3','a4'],
             ['b0','b1','b2','b3','b4']])
ar
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s')
```

```python
y={'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',
            annot=True,annot_kws=y)
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',
            annot=ar,fmt='s',
            linewidths=10,linecolor='y')
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',
            annot=ar,fmt='s',
            linewidths=10,linecolor='y',
            cbar=False)
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',
            annot=ar,fmt='s',
            linewidths=10,linecolor='y',
            cbar=False,xticklabels=False)
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',
            annot=ar,fmt='s',
            linewidths=10,linecolor='y',
            cbar=False,yticklabels=False,xticklabels=True)
```

```python
v=sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',
            annot=ar,fmt='s',
            linewidths=10,linecolor='y',
            cbar=False,yticklabels=False,xticklabels=True)
v.set(xlabel='Python',ylabel="Y-axis")
plt.show()
```

```python
v=sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',
            annot=ar,fmt='s',
            linewidths=10,linecolor='y',
            cbar=False,yticklabels=False,xticklabels=True)
v.set(xlabel='Python',ylabel="Y-axis")
sns.set(font_scale=5)
plt.show()
```

# Count Plot in seaborn

```python
var3=sns.load_dataset('tips')
var3
```

```python
sns.countplot(x='sex',data=var3)
plt.show()
```

```python
sns.barplot(x='sex',y='size',data=var3)
```

```python
sns.barplot(x='sex',y='size',data=var3,hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex',data=var3,hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex',
              data=var3,hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex',data=var3,
              hue='smoker',color='r')
plt.show()
```

```python
sns.countplot(x='sex',data=var3,
              hue='smoker',color='r',saturation=.5)
plt.show()
```
