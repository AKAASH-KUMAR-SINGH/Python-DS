# Bar Plot In Seaborn

```python
y_2=sns.load_dataset('penguins')
y_2
```

```python
sns.barplot(x=y_2.island,y=y_2.bill_length_mm)
plt.show()
```

```python
# Without using ".y_2" name 
sns.barplot(x="island",y='bill_length_mm',data=y_2)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=y_2,hue='sex')
plt.show()
```

```python
order_1=['Dream','Biscoe',"Torgersen"]
sns.barplot(x='island',y='bill_length_mm',data=y_2,hue='sex',order=order_1)
plt.show()
```

```python
order_1=['Dream','Biscoe',"Torgersen"]
sns.barplot(x='island',y='bill_length_mm',data=y_2,
            hue='sex',order=order_1,hue_order=['Female','Male'])
plt.show()
```

```python
order_1=['Dream','Biscoe',"Torgersen"]
sns.barplot(x='island',y='bill_length_mm',data=y_2,hue='sex',
            order=order_1,hue_order=['Female','Male'],ci=100)
plt.show()
```

```python
order_1=['Dream','Biscoe',"Torgersen"]
sns.barplot(x='island',y='bill_length_mm',data=y_2,hue='sex',
            order=order_1,hue_order=['Female','Male'],ci=100,
            n_boot=2)
plt.show()
```

```python
order_1=['Dream','Biscoe',"Torgersen"]
sns.barplot(x='island',y='bill_length_mm',data=y_2,hue='sex',
            order=order_1,hue_order=['Female','Male'],orient='v',
            palette='icefire')
plt.show()
```
