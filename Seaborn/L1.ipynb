```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
var=[1,2,3,4,5,6,7]
var_1=[3,2,4,3,5,1,4]
plt.plot(var,var_1)
plt.show()
```

```python
v_1=pd.DataFrame({'var':var,'var_1':var_1})
sns.lineplot(x='var',y='var_1',data=v_1)
plt.show()
```

```python
y_1=sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y_1)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=y_1,hue='sex',size=50)
plt.show()
```

```python
sns.lineplot(x='bill_depth_mm',y='bill_length_mm',data=y_1,hue='sex',style='sex',palette='Accent',
             markers=['o','>'],dashes=True,legend='full')
plt.grid()
plt.title("python")
plt.show()
```
