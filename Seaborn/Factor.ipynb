# Cat Plot In Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset('tips').head(20)
var
```

```python
sns.catplot(x='tip',y='size',data=var)
plt.show()
```

```python
sns.catplot(x='tip',y='size',data=var,hue='sex')
```

```python
sns.catplot(x='tip',y='size',data=var,hue='sex',palette='Oranges')
```

```python
sns.catplot(x='tip',y='size',data=var,hue='sex',palette='Oranges',kind='bar')
```

```python
sns.catplot(x='tip',y='size',data=var,hue='sex',palette='Oranges',kind='box')
```

```python
sns.catplot(x='tip',y='size',data=var,hue='sex',palette='Oranges',kind='box')
```

# Styling Plots In Seaborn

Seaborn Figure Style
Removing Axes Spines
Scale and Context

```python
var
```

```python
sns.boxplot(y=var['total_bill'])
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=var)
plt.grid()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.barplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(5,5))
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.show()
```

```python
sns.set_style('white')
sns.set_context('poster',font_scale=.5)
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.show()
```

# Multiple Plot(Facet-Grid) In Seaborn

```python
var=sns.load_dataset('tips')
var
```

```python
fg=sns.FacetGrid(var)
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg=sns.FacetGrid(var,col='sex')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg=sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg=sns.FacetGrid(var,col='sex',hue='day')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col='day',hue='sex')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col='day',hue='sex')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```
